```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Setting random seed for reproducibility
np.random.seed(42)

# Generate synthetic ATM withdrawal amounts
mean = 2000
std_dev = 500
data = np.random.normal(mean, std_dev, 100)
```

```python
# Add some outliers
outliers = np.array([3500, 4000, 4500, 5000])
data_with_outliers = np.concatenate([data, outliers])
```

```python
# First Chart: Normal distribution with standard deviation markers
plt.figure(figsize=(10, 6))
sns.histplot(data_with_outliers, bins=20, kde=True, color='blue', stat='density')
# Plot standard deviation lines
for sd in range(-3, 4):
    value = mean + sd * std_dev
    plt.axvline(value, linestyle='--', linewidth=1, color='red' if sd != 0 else 'green')
    label = f'{sd}Ïƒ\n{value}' if sd != 0 else f'Mean\n{value}'
    plt.text(value, plt.ylim()[1]*0.9, label, horizontalalignment='center', color='red' if sd != 0 else 'green')

plt.title('ATM Withdrawal Amounts with Standard Deviations')
plt.xlabel('Withdrawal Amount ($)')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# Second Chart: Boxplot with Percentiles
plt.figure(figsize=(10, 4))
sns.boxplot(x=data_with_outliers, color='skyblue')

# Annotate percentiles
percentiles = [0, 25, 50, 75, 100]
percentile_values = np.percentile(data_with_outliers, percentiles)



plt.title('Boxplot of ATM Withdrawal Amounts with Percentiles')
plt.xlabel('Withdrawal Amount ($)')
plt.grid(True)
plt.show()
```
